## Task 2: Equal Oppertunities
According to its social policies on equal opportunities, 4-Airlines decided that women should give at
least 40% of the total number of employees. In this context, the gender-based structure of the company’s employees for the last five years is disclosed in Table 4 (included in the company’s sustainability
report).

Data table:

```python
import pandas as pd
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Total employees': [103764, 82549, 78967, 82132, 75507],
    'Female employees': [47731, 34670, 33956, 34495, 29448]
}
df = pd.DataFrame(data)
print(df.to_string(index=False))
```

### Our Task
Prepare an audit work sheet in order to test the management assertion in this matter, based
on data disclosed in Table 4
Considering the outcome of this task, as well as the fact that the management intends to
disclose only absolute numbers (no relative numbers) regarding its employees in the sustainability report, determine the impact of this presentation on the audit opinion.

```python
df['Female percentage'] = (df['Female employees'] / df['Total employees']) * 100
df['Female percentage'] = df['Female percentage'].round(2)
print("Table 4 - Total number of employees and percentage of female employees (2019-2023)")
print(df.to_string(index=False))
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6)) 
plt.plot(df['Year'], df['Female percentage'], 
         marker='o',  
         linestyle='-',  
         color='blue',  
         linewidth=2)  
REF_LINE = 40  
plt.axhline(y=REF_LINE, color='red', linestyle='--', linewidth=1.5)
plt.xticks(df['Year'])
plt.title('Percentage of Female Employees (2019–2023)', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Female Employees (%)', fontsize=12)
plt.ylim(35, 50) 
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

It is clear that the number of female employes has been in demise over the years. The company's goal of 40% women employment was achieved every year till 2023. This can of course have impact on our assurance opininion if the company is declaring a 40% women employment, because this number is not true. If the company is only saying that 40% is the goal, our audit opinion would be unqualified.

## Task 4 Supply Chain
According to the business ethics policy of 4-Airlines on supporting domestic markets, the company 
should source at least 63% of the total value of its purchases from local suppliers. For the period 2019 – 2023, purchases were structured as follows:  

```python
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Total Purchases in €': [892, 720, 831, 927, 1148],
    "Domestic Purchases in €": [564, 480, 536, 612, 717]
}
df = pd.DataFrame(data)
print(df.to_string(index=False))
```

Materials in total amount of 32.5 million euro were purchased according to DAP INCOTERMS condi
tions. The invoice is dated 12/29/2023 by a supplier located in Spain, while the goods were received in 
Vienna, as location indicated in the purchase documents, on 1/5/2024. The company presented this 
acquisition as related to the year 2023. 
### Our Task:
Describe the impact of this information on the Sustainability assurance report and make a 
disclosure adjustment proposal for the management.

Let's start by first checking how much of the goods were sources locally in previous years.

```python
df["Domestic Percentage"] = df["Domestic Purchases in €"] / df["Total Purchases in €"] * 100
print(df["Domestic Percentage"])
```

We see that companys goal of at least 63% domestic purchases was achieved almost every year, with 2023 being an exeption. Now let's analyse how the latest findings about DAP INCOTERMS impact the sustainability statement. In order to do this we need to exclude the 32.5 million € from the total purchases column and leave the domestic purchases column unchanged. 

```python
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Total Purchases in €': [892, 720, 831, 927, 1148 - 32.5],
    "Domestic Purchases in €": [564, 480, 536, 612, 717]
}
df = pd.DataFrame(data)
df["Domestic Percentage"] = df["Domestic Purchases in €"] / df["Total Purchases in €"] * 100
print(df.to_string(index=False))
```

This small calculation effectivly changes the percentage of domestic purchases, pushing it above diseried 63%, meaning the company has achieved it's goal
